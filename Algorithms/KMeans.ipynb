```python
import json
import random
import math
from pprint import pprint
```

```python
k = 3
location_list = []
centroid_list = []
group_dict = {}

with open('file.json', 'r') as f:
    current_json = json.load(f)
    info_list = current_json
    for info in info_list:
        location_list.append((info['coordinates']['lat'], info['coordinates']['lng']))

for i in range(k):
    random_point = random.choice(location_list)
    while random_point in centroid_list:
        random_point = random.choice(location_list)
    centroid_list.append(random_point)

is_running = True
iteration_count = 0

while is_running:
    for location in location_list:
        best_group = -1
        best_distance = 100000000000
        location_x = location[0]
        location_y = location[1]
        for index, centroid in enumerate(centroid_list):
            centroid_x = centroid[0]
            centroid_y = centroid[1]
            distance_cost = math.sqrt(((centroid_x - location_x) ** 2) + ((centroid_y - location_y) ** 2))
            if distance_cost < best_distance:
                best_group = index
                best_distance = distance_cost
        group_dict[location] = best_group

    update_dict = {}
    for location in group_dict.keys():
        if group_dict[location] not in update_dict:
            update_dict[group_dict[location]] = []
        update_dict[group_dict[location]].append(location)
    
    is_constant = True
    
    for group_id in update_dict.keys():
        sum_x = sum([x for x, y in update_dict[group_id]])
        sum_y = sum([y for x, y in update_dict[group_id]])
        
        new_x = sum_x / len(update_dict[group_id])
        new_y = sum_y / len(update_dict[group_id])
        
        centroid_x = centroid_list[group_id][0]
        centroid_y = centroid_list[group_id][1]
        
        difference_cost = math.sqrt(((centroid_x - new_x) ** 2) + ((centroid_y - new_y) ** 2))
        
        if difference_cost > 0.00001:
            centroid_list[group_id] = (new_x, new_y)
            is_constant = False
    
    if is_constant:
        is_running = False
    
    iteration_count += 1

print("Points -> Groups")
pprint(group_dict)
print("Centroids")
pprint(centroid_list)
```
